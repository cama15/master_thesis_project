```python
import gzip

def parse_line(line):
    fields = line.strip().split('\t')
    chrom, start, end = fields[0], int(fields[1]), int(fields[2])
    meth_level = float(fields[3])
    coverage = int(fields[4])
    return chrom, start, end, meth_level, coverage

def read_bed_file(filename):
    merged = {}
    open_func = gzip.open if filename.endswith(".gz") else open
    with open_func(filename, 'rt') as f:
        for line in f:
            if line.startswith("#") or line.strip() == "":
                continue
            chrom, start, end, meth_level, coverage = parse_line(line)
            key = (chrom, start, end)
            if key not in merged:
                merged[key] = {'meth_level_sum': 0.0, 'coverage_sum': 0}
            merged[key]['meth_level_sum'] += meth_level * coverage
            merged[key]['coverage_sum'] += coverage
    return merged

def merge_dicts(dict1, dict2):
    for key, vals in dict2.items():
        if key in dict1:
            dict1[key]['meth_level_sum'] += vals['meth_level_sum']
            dict1[key]['coverage_sum'] += vals['coverage_sum']
        else:
            dict1[key] = vals
    return dict1

def write_merged_bed(merged_dict, output_file):
    with open(output_file, 'w') as out:
        for (chrom, start, end), vals in sorted(merged_dict.items()):
            cov = vals['coverage_sum']
            avg_meth = (vals['meth_level_sum'] / cov) if cov > 0 else 0.0
            out.write(f"{chrom}\t{start}\t{end}\t{avg_meth:.3f}\t{cov}\n")

if __name__ == "__main__":
    lane1_file = "ICM_lane1_Homo_sapiens_Bisulfite-Seq_mergecg.bed.gz"
    lane2_file = "ICM_lane2_Homo_sapiens_Bisulfite-Seq_mergecg.bed.gz"
    output_file = "ICM_lane1+2_merged_mergecg.bed"

    print("Reading lane 1 file...")
    lane1_data = read_bed_file(lane1_file)

    print("Reading lane 2 file...")
    lane2_data = read_bed_file(lane2_file)

    print("Merging data...")
    merged_data = merge_dicts(lane1_data, lane2_data)

    print(f"Writing merged data to {output_file} ...")
    write_merged_bed(merged_data, output_file)

    print("Done!")
```

```python
from collections import defaultdict

def count_high_coverage_sites(file_path, threshold=10):
    chrom_high_cov_counts = defaultdict(int)
    total_high_cov = 0
    
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith("#") or line.strip() == "":
                continue
            fields = line.strip().split('\t')
            chrom = fields[0]
            coverage = int(fields[4])
            
            if coverage >= threshold:
                chrom_high_cov_counts[chrom] += 1
                total_high_cov += 1
    
    print(f"Total CpG sites with coverage ≥ {threshold}: {total_high_cov}\n")
    print(f"Breakdown by chromosome:")
    for chrom in sorted(chrom_high_cov_counts.keys()):
        print(f"{chrom}: {chrom_high_cov_counts[chrom]}")

if __name__ == "__main__":
    merged_file = "ICM_lane1+2_merged_mergecg.bed"  # Replace with the actual path if needed
    count_high_coverage_sites(merged_file, threshold=5)
```

```python
def filter_by_coverage(input_file, output_file, min_cov=5):
    with open(input_file, 'r') as fin, open(output_file, 'w') as fout:
        for line in fin:
            fields = line.strip().split('\t')
            coverage = int(fields[4])
            if coverage >= min_cov:
                fout.write(line)
```
