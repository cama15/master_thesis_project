```python
import pandas as pd
```

```python
df1 = pd.read_csv("samples_1_matrix.csv", sep=",")
df2 = pd.read_csv("samples_2_matrix.csv", sep=",")
```

```python
print(df1.head(10))
print(df2.head(10))
```

```python
df1["Chr"] = df1["Chr"].astype(str)
df2["Chr"] = df2["Chr"].astype(str)
```

```python
df1 = df1.rename(columns={"Chr": "Chromosome"})
df2 = df2.rename(columns={"Chr": "Chromosome"})
```

```python
merged_df = pd.merge(df1, df2, on=["Chromosome", "Start", "End"], how="outer")
```

```python
print(merged_df.tail(20))
```

```python
print(merged_df.tail(45))
```

```python
merged_df.to_csv("meth_matrix_10_samples.csv", sep=",", index=False)
```
